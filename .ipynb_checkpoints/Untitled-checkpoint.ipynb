```python
def to_if(name):
     if name == "Jan": 
            return 1
     elif name == "Feb": 
            return 2
     elif name == "Mar": 
            return 3
     elif name == "Apr": 
            return 4
     elif name == "May": 
            return 5
     elif name == "Jun": 
            return 6
     elif name == "Jul": 
            return 7
     elif name == "Aug": 
            return 8
     elif name == "Sep": 
            return 9
     elif name == "Oct": 
            return 10
     elif name == "Nov": 
            return 11
     elif name == "Dec": 
            return 12
     else: raise ValueError
```

```python
import pandas as pd
import calendar

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("Dataset.csv")
```

```python
print(df)
```

```python
#Values with binary data
df['Sex'] = df['Sex'].apply(lambda x: 0 if x=='Male' else 1)#sex(male=0/female=1)
df['Fault'] = df['Fault'].apply(lambda x: 0 if x=='Policy Holder' else 1)#fault(Policy holder=0/Third party=1)
df['AccidentArea'] = df['AccidentArea'].apply(lambda x: 0 if x=='Urban' else 1)#Accident area(urban=0/rural=1)
df['PoliceReportFiled'] = df['PoliceReportFiled'].apply(lambda x: 0 if x=='No' else 1)#Police report filed(Yes=1/no=0)
df['WitnessPresent'] = df['WitnessPresent'].apply(lambda x: 0 if x=='No' else 1)#Witness present(Yes=1/No=0)
```

```python
#months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
#df["Month"]= months.index(df["Month"])
df['Month'] = df['Month'].apply(lambda x: calendar.month_abbr[x])
print(df["Month"])
```

```python
#number of cars
df['NumberOfCars']=df['NumberOfCars'].str.rstrip(' vehicle').str.rstrip(' vehicles')
print(df)
```

```python
y = df["FraudFound_P"] #creating a dataframe which can store the output columns
X = df.drop(columns=['FraudFound_P','PolicyNumber']) #deleating or droping the result colum from the given data set and storing it in a new data frame X
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33) #splitting the train and test data with 33% of total data as test data
```

```python
#Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train) #plotting for prediction
y_train_pred = lr.predict(X_train) #predicting and checking the correctness of the train data in the drawn plot
y_test_pred = lr.predict(X_test) #predicting and checking the correctness of the test data in the drawn plot
print(sum(y_test_pred == y_test) / len(y_test)) #Checking the probability of correct resullt on test data
print(sum(y_train_pred == y_train) / len(y_train))  #Checking the probability of correct resullt on train data
```
